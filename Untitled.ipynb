# Prepare data - extract columns

Remain only:
    Pclass;
    Sex;
    Age;
    Number of parents (adult) in train;
    Number of survived parents (adult) in train.

```python
import math
```

```python
columns_in_test_data = ["Pclass", "Sex", "Age"]
columns = ["Pclass", "Sex", "Age", "ParInTrain", "ParInTrainSurvived"]
```

```python
any(x in "Todoroff, Mr. Lalio" for x in ["Mr", "Mrs"])
```

```python
prepared_train_data = train_data[columns_in_test_data]

par_in_train = []
par_in_train_surv = []

for i in range(len(prepared_train_data)):
    if prepared_train_data.at[i, "Sex"] == "male":
        prepared_train_data.at[i, "Sex"] = 1
    else:
        prepared_train_data.at[i, "Sex"] = 0
        
    if math.isnan(prepared_train_data.at[i, "Age"]):
        if any(x in train_data.at[i, "Name"] for x in ["Mr", "Mrs"]):
            prepared_train_data.at[i, "Age"] = 30
        else:
            prepared_train_data.at[i, "Age"] = 12
            
    pit = 0
    pits = 0
    if train_data.at[i, "Age"] <= 18:
        surname = train_data.at[i, "Name"].split(',')[0]
        for j in range(len(train_data)):
            if i == j:
                continue
            if surname in train_data.at[j, "Name"] and train_data.at[j, "Age"] > 18:
                pit += 1
                if train_data.at[j, "Survived"] == 1:
                    pits += 1
                
    par_in_train.append(pit)
    par_in_train_surv.append(pits)
        
prepared_train_data["Par_in_train"] = par_in_train
prepared_train_data["Par_in_train_surv"] = par_in_train_surv
prepared_train_data.head(30)
```

```python
y_train_data = train_data["Survived"]
y_train_data.head()
```

```python
prepared_test_data = test_data[columns_in_test_data]

par_in_train = []
par_in_train_surv = []

for i in range(len(prepared_test_data)):
    if prepared_test_data.at[i, "Sex"] == "male":
        prepared_test_data.at[i, "Sex"] = 1
    else:
        prepared_test_data.at[i, "Sex"] = 0
        
    if math.isnan(prepared_test_data.at[i, "Age"]):
        if any(x in test_data.at[i, "Name"] for x in ["Mr", "Mrs"]):
            prepared_test_data.at[i, "Age"] = 30
        else:
            prepared_test_data.at[i, "Age"] = 12
            
    pit = 0
    pits = 0
    if test_data.at[i, "Age"] <= 18:
        surname = test_data.at[i, "Name"].split(',')[0]
        for j in range(len(train_data)):
            if i == j:
                continue
            if surname in train_data.at[j, "Name"] and train_data.at[j, "Age"] > 18:
                pit += 1
                if train_data.at[j, "Survived"] == 1:
                    pits += 1
                
    par_in_train.append(pit)
    par_in_train_surv.append(pits)
        
prepared_test_data["Par_in_train"] = par_in_train
prepared_test_data["Par_in_train_surv"] = par_in_train_surv
        
prepared_test_data.head(30)
```

```python
from sklearn.svm import SVC

#model = DecisionTreeClassifier(random_state = 1)
svc = SVC(kernel='rbf', C=10, gamma=1)

svc.fit(prepared_train_data, y_train_data)

predict_test = svc.predict(prepared_test_data)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import RandomForestClassifier

#model = DecisionTreeClassifier(random_state = 1)
model = RandomForestClassifier(max_depth = 5, random_state = 1)

model.fit(prepared_train_data, y_train_data)

predict_test = model.predict(prepared_test_data)
```

```python
output = pd.DataFrame(list(zip(range(892, 892+len(predict_test)), predict_test)), 
                      columns = ['PassengerId', 'Survived'])
output.head()
```
